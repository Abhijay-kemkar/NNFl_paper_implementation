```python
#imports
import cv2, os
import matplotlib.image as mpimg 
import matplotlib.pyplot as plt 
import numpy as np
from PIL import Image
```

```python
def load_from_single_page_tiff(path_to_image):
    """
    Convert image to array
    :param path_to_image:
    :return: np_array
    """
    return np.array(Image.open(path_to_image))
```

```python
#base_path = "01_ST/SEG/"
base_path = "02_ST/SEG/"
new_path = "data/y/"
for infile in os.listdir(base_path):                                            #iterate though the tiff files
    print ("file : " + infile)
    read = load_from_single_page_tiff(base_path + infile)                       #loading tiffs and returning np arrays
    #outfile = infile.split('.')[0] + '1' + '.png'           
    outfile = infile.split('.')[0] + '2' + '.png'                               #changing naming structure as there are 2 folders of images 
    cv2.imwrite(new_path+outfile,read,[int(cv2.IMWRITE_JPEG_QUALITY), 200])     #creating png from arrays
```
